```python
from time import sleep

print('''
Farei algumas perguntas e as respostas serão de multipla escolha.
Marque a opção que você julga correta e teste os seus conhecimentos.
''')

sleep(1.5)

perguntas = {
    'pergunta1': {
        'pergunta': 'Quanto é 2+2? ',
        'resposta': {
            'a': '1',
            'b': '4',
            'c': '5',
            'd': '2',
        },
        'resposta_certa': 'b',
    },
    'pergunta2': {
        'pergunta': 'Quanto é 5*5? ',
        'resposta': {
            'a': '10',
            'b': '35',
            'c': '25',
            'd': '15',
        },
        'resposta_certa': 'c',
    },
    'pergunta3': {
        'pergunta': 'Quanto é 1+2? ',
        'resposta': {
            'a': '1',
            'b': '4',
            'c': '2',
            'd': '3',
        },
        'resposta_certa': 'd', 
    },
    'pergunta4': {
        'pergunta': 'Quanto é 1-1? ',
        'resposta': {
            'a': '0',
            'b': '5',
            'c': '1',
            'd': '2',
        },
        'resposta_certa': 'a',
    },
    'pergunta5': {
        'pergunta': 'Quanto é 8/4? ',
        'resposta': {
            'a': '4',
            'b': '16',
            'c': '12',
            'd': '2',
        },
        'resposta_certa': 'd',
    },      
}

respostas_certas = 0
for pk, pv in perguntas.items():
  print(f'{pk}: {pv["pergunta"]}')

  print('Respostas: ')
  for rk, rv in pv['resposta'].items():
    print(f'[{rk}]: {rv}')

  resposta_usuario = input('Qual a sua resposta? ')
  if resposta_usuario == pv['resposta_certa']:
    sleep(0.5)
    print('Você acertou. Parabéns!')
    print(f'Resposta correta: {pv["resposta_certa"]}')
    respostas_certas += 1
  else:
    sleep(0.5)
    print('Infelizmente você errou.')
    print(f'Resposta correta: {pv["resposta_certa"]}')

  print()

sleep(1.5)
qnt_perguntas = len(perguntas)
porcentagem_acertos = respostas_certas / qnt_perguntas * 100

print(f'Você acertou {respostas_certas} respostas.')
print(f'Seu percentual de acertos foi de {porcentagem_acertos}%.')
```
